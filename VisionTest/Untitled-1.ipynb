```python
# 감지된 물체의 실제 물체의 너비와 카메라의 초점 거리 그리고 바운딩 박스의 너비를 사용하여 물체와의 거리를 계산한다.

from ultralytics import YOLO
import cv2

# LCOCO 데이터셋으로 사전 훈련된 YOLO 모델을 로드합니다.
model = YOLO("best0119.pt")

# 웹캠 열기
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠을 열 수 없습니다.")
    exit()

# 실제 물체의 크기 (cm)
actual_width = 10  # 예: 물체의 실제 너비가 10cm라고 가정
focal_length = 700.0  # 카메라의 초점 거리 (픽셀 단위, 실험을 통해 결정)

while True:
    # 프레임 읽기
    ret, frame = cap.read()
    if not ret:
        print("프레임을 읽을 수 없습니다.")
        break
    
    # YOLO 모델을 사용하여 물체 감지
    results = model(frame)

    # 결과를 이미지에 그리기
    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]  # 바운딩 박스 좌표
            width = x2 - x1  # 바운딩 박스의 너비
            class_id = int(box.cls[0])  # 클래스 ID
            class_name = model.names[class_id]  # 클래스 이름

            # 거리 계산 (cm)
            distance = (actual_width * focal_length) / width

            # 바운딩 박스 그리기
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)

            # 텍스트 배경 그리기
            text = f'{class_name} / Distance: {distance:.2f} cm'
            text_size = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)[0]
            text_x = int(x1)
            text_y = int(y1) - 10
            
            # 텍스트 배경 사각형 그리기
            cv2.rectangle(frame, (text_x, text_y - text_size[1]), (text_x + text_size[0], text_y + 5), (255, 0, 0), -1)
            
            # 텍스트 그리기
            cv2.putText(frame, text, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    # 프레임 표시
    cv2.imshow('Webcam', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 웹캠 해제 및 모든 창 닫기
cap.release()
cv2.destroyAllWindows()
```
